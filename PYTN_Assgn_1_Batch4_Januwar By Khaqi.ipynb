### Menampilkan Major_Category = Robbery dari Tahun 2008 - 2016

```python
major = df[(df['year'] > 2007)].groupby(['major_category','year'])['value'].sum()
robbery = major['Robbery']
robbery
df_robbery = robbery.to_frame()
df_robbery = df_robbery.rename(columns={'value': 'Robbery'})
df_robbery
```

### Menampilkan Major_Category = Burglary dari Tahun 2008 - 2016

```python
burglary = major['Burglary']
burglary
df_burglary = burglary.to_frame()
df_burglary = df_burglary.rename(columns={'value': 'Burglary'})
df_burglary
```

### Menggabungkan 2 data frame

```python
df_burglary_robbery = pd.concat([df_robbery, df_burglary], axis=1)
df_burglary_robbery
```

### Menampilkan Area Plot perbandingan Burlgary dan Robbery dari Tahun 2008 - 2016

<span style="color:darkblue">Terlihat data diatas bahwa hasil compare dari Burglary vs Robbery, selama tahun 2008 - 2016 Burglary masih jadi trend kasus crime di london</span>.

### Membuat Data Frame Lonngitude, Latitude dari Top 5 Crime Tertinggi

```python
country_dict = {
    'country': ['Westminster', 'Southwark', 'Lambeth', 'Camden', 'Newham'],
    'Latitude': [ 51.5002, 51.5014, 51.4995, 51.5517, 51.5255],
    'Longitude': [0.1262, 0.1047, 0.1117, 0.1588, 0.1588 ]
}

df_country_longlat = pd.DataFrame(country_dict)
df_country_longlat
```

### Menandai 5 Country pada Foluim Map

```python
incidents = folium.map.FeatureGroup()
for lat, lng, in zip(df_country_longlat.Latitude, df_country_longlat.Longitude):
    incidents.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=10, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )
```

```python
region.add_child(incidents)
```
